Actuarial notation 精算符号    

present value（PV）：
$$
\begin{array}{l}{\text { Present Value }=\frac{\mathrm{FV}}{(1+r)^{n}}} \\ {\text { where: }} \\ {\mathrm{FV}=\text { Future Value }} \\ {r=\text { Rate of return }} \\ {n=\text { Number of periods }}\end{array}
$$

consol(永续年金) or perpetuity(公债)：
$$
P V=\sum_{k=1}^{\infty} \frac{C}{(1+i)^{k}}=\frac{C}{i}, \quad i>0
$$
永续年金其实就就是在年金的的公式上求极限  
市场利率提高后，公债的PV价值降低

growing consol(例如每次支付增长g%）:
$$
P V=\frac{C}{i-g}, \quad i>g
$$

annuity（年金）:
$$
P V=C \frac{1-(1 /(1+r))^{n}}{r}
$$

保险与二项式分布  

几何平均估算平均收益率比算数平均来的更好, 为什么呢？  
$$ 
\left(\prod_{i=1}^{n} x_{i}\right)^{\frac{1}{n}}=\sqrt[n]{x_{1} x_{2} \cdots x_{n}}
$$

期望值(金融上称为预期收益率Expected Return)：
$$
\mu=\sum_{i=1}^{n} p_{i} \cdot x_{i}
$$  

方差 
$$
\operatorname{Var}(X)=\sum_{i=1}^{n} p_{i} \cdot\left(x_{i}-\mu\right)^{2}=\sum_{i=1}^{n}\left(p_{i} \cdot x_{i}^{2}\right)-\mu^{2}
$$

样本方差：
$$
s^{2}=\frac{1}{n-1} \sum_{i=1}^{n}\left(X_{i}-\bar{X}\right)^{2}
$$

协方差：
$$
\operatorname{cov}(X, Y)=\sum_{i=1}^{N} \frac{\left(x_{i}-\bar{x}\right)\left(y_{i}-\bar{y}\right)}{N}
$$

correlation(相关系数)：
$$
\rho_{X, Y}=\operatorname{corr}(X, Y)=\frac{\operatorname{cov}(X, Y)}{\sigma_{X} \sigma_{Y}}=\frac{\mathrm{E}\left[\left(X-\mu_{X}\right)\left(Y-\mu_{Y}\right)\right]}{\sigma_{X} \sigma_{Y}}
$$
